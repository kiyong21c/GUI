```
import pandas as pd
import requests
from bs4 import BeautifulSoup

date = list(range(20210101, 20210110))

titles = []
points = []

for time in date:
  url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={}".format(date)
  res = requests.get(url)
  res.raise_for_status()  # 문제있으면 프로그램 종료
  soup = BeautifulSoup(res.text, "lxml")


  title_list = soup.find_all("td", attrs={"class":"title"})
  for title in title_list:
    titles.append(title.a.get_text())

  point_list = soup.find_all("td", attrs={"class":"point"})
  for point in point_list:
    points.append(point.get_text())


df = pd.DataFrame({'name':titles, 'point': points})
df
```
