```
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

```

```
# dates = list(range(20211210, 20211210))
dates = [20211220, 20211221] # '문자열{}'.format(int형)을 통해 int형이 문자열로 들어감
titles = []
when = []

for date in dates:
  url = "https://finance.naver.com/news/news_list.naver?mode=LSS3D&section_id=101&section_id2=258&section_id3=403&date={}".format(date)
  headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.232 Whale/2.10.124.26 Safari/537.36"}
  res = requests.get(url, headers=headers)
  res.raise_for_status()  # 문제있으면 프로그램 종료
  soup = bs(res.text, "lxml")

# find_all 보다 select를 활용하는 방식이 더욱 자유도가 높고 편하다.

  title_list = soup.select(".articleSubject") # 태그는 모르겠고, 클래스로 모두 선택할때 : select('.클래스')
  for title in title_list:
    if title:
      titles.append(title.a.get_text())
    else:
      titles.append('-')

  wdate_list = soup.select('dd.articleSummary > span.wdate') # select('태그.클래스 > 하위태그.클래스')
  for wdate in wdate_list:
    if wdate:
      when.append(wdate.get_text())
    else:
      when.append('-')

df = pd.DataFrame({'제목':titles, '작성일':when}) # 컬럼별 데이터의 개수가 안맞으면 데이터 프레임 구성 안됨
df
```

# 문자열 형태인 날짜를 수정
  - object -> datetime -> object

```
df['작성일'].head()
```

```
pd.to_datetime(df['작성일']).head() # pd.to_datetime()으로 series를 datetime형태로 변경
```

```
df['작성일'] = pd.to_datetime(df['작성일'])
df.head()
```

```
df['작성일'].dt.strftime('%y-%m-%d').head() # datetime형태의 series를 수정하는 dt함수를 사용하여 strftime()함수로 형식변경('원하는 문자열형태')
```

```
df['작성일'] = df['작성일'].dt.strftime('%y-%m-%d')
df['작성일'].head()
```
