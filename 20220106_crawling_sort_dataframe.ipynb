```
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests

date = input("날짜를 입력하세요(ex:20220101): ")
url = "https://finance.naver.com/news/news_list.naver?mode=RANK&date={}".format(date)

res = requests.get(url)
soup = bs(res.text, 'lxml')

items = soup.select('li.block1.clearfix li')  # 클래스명 사이에 띄워쓰기가 있으면 점(.) 찍자

title = []
press = []
wdate = []
for item in items:
  title.append(item.a.get_text())
  press.append(item.select_one('span.press').get_text())
  wdate.append(item.select_one('span.wdate').get_text())

df = pd.DataFrame({"제목":title, "언론사":press, "작성일":wdate}) # 데이터 프레임
df['작성일'] = pd.to_datetime(df['작성일']) # 작성일 컬럼 Datatime 객체로 변환
df = df.sort_values('언론사').set_index('언론사') # 언론사별 그룹화하여 인덱스 설정
df
```
