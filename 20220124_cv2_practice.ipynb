```
import numpy as np

# 색 사용을 위한 변수 설정(튜플)
blue_color = (255, 0, 0)
green_color = (0, 255, 0)
red_color = (0, 0, 255)
white_color = (255, 255, 255)

# 두께 변수 설정
width = 3

# 모두 0으로 되어 있는 빈 Canvas(검정색)
img = np.zeros((384, 384, 3), np.uint8)

# 이미지 그리기
img = cv2.rectangle(img, (10, 10), (100, 100), blue_color, width)
img = cv2.rectangle(img, (110, 110), (200, 200), green_color, width)
img = cv2.rectangle(img, (210, 210), (350, 350), red_color, width)

cv2_imshow(img)
```
