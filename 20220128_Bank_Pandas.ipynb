```
import pandas as pd
```

```
df = pd.read_excel('거래내역.xls', skiprows = 7, usecols=[2,3,4,5,6,7])
df = df.fillna('')
```

```
df
```

```
df_credit = df[df['거래내용']=="카드대금"]
```

```
df_credit
```

```
df_credit[df_credit['거래기록사항']=='삼성카드']
```

```
df[df['거래내용'].str.contains('보험', na=False)]   # 시리즈에 문자열 함수를 사용하고 싶다면 Series.str.문자열함수()
# na = False 처리하지 않으면 에러 : Cannot mask with non-boolean array containing NA / NaN values
# True/False 로 구분되지 않는 사항(NaN 등)에 대한 처리를 하지 못하기 때문에 공백은 False 처리해야함
```

```
df_insure = df[df['거래내용'].str.contains('보험', na=False)]
df_insure
```
