```
import pandas as pd
df = pd.DataFrame({'날짜':pd.date_range('2021-12-16', periods=100, freq='3D')})
df.head()
```

```
# 매년 15일과 같은 특정 날짜 datetime periods 만들기(OT 기준일)
pd.date_range('2021-12-1', periods=13, freq='MS') + timedelta(days=14)
# freq='MS' : 매월(Month)초(Start)
# 매월초 + timedelta → 매월 특정일 설정(다른방법 없음)
```

```
from datetime import datetime, timedelta
dt = pd.date_range('2021-12-1', periods=13, freq='MS') + timedelta(days=14)
for i in range(0, 12):
    df.loc[(df['날짜'] <= dt[i+1]) & (df['날짜'] > dt[i]), '해당월'] = str(dt[i+1].month) + "월"
    # for문 통해 i+1은 12까지 적용되므로 12번째 인덱스를 사용하기 위해 dt 리스트는 13개의 값이 필요함 
df
```
