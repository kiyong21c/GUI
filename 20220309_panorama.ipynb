```
import cv2
import numpy as np

imgL = cv2.imread('/content/drive/MyDrive/Colab Notebooks/restaurant1.jpg')
imgR = cv2.imread('/content/drive/MyDrive/Colab Notebooks/restaurant2.jpg')

hL, wL = imgL.shape[:2]
hR, wR = imgR.shape[:2]

grayL = cv2.cvtColor(imgL, cv2.COLOR_BGR2GRAY)
grayR = cv2.cvtColor(imgR, cv2.COLOR_BGR2GRAY)

# ORB 추출기 생성
orb = cv2.ORB_create()
# 키포인트 검출 및 디스크립터 계산
kpsL, descL = orb.detectAndCompute(imgL, None)
kpsR, descR = orb.detectAndCompute(imgR, None)

#BF매칭기 생성 및 knn 매칭
matcher = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=False)
matches = matcher.knnMatch(descL, descR, 2)

# match(), knnMatch(), radiusMatch()함수의 반환결과는 Dmatch 객체가 담긴 리스트
# 첫번째 이웃의 거리가 두번째 이웃 거리의 75% 이내인 것만 추출
ratio = 0.75
good_matches = [first for first, second in matches if first.distance < second.distance * ratio]
print("총 good_matches :", len(good_matches))

if len(good_matches) > 4:
  ptsL = np.float32([kpsL[m.queryIdx].pt for m in good_matches])
  ptsR = np.float32([kpsR[m.trainIdx].pt for m in good_matches])
  # 원근 변환 행렬 구하기(좋은 매칭점들로 부터 근사계산한) : 변환 행렬의 shape (3, 3)
  mtrx, mask = cv2.findHomography(ptsL, ptsR)
  # 원근 변환 행렬 적용해서 오른쪽 사진 원근 변환, 결과 이미지 크기는 사진 2장 크기
  panorama = cv2.warpPerspective(imgR, mtrx, (wL+wR, hR))


# print(matches)
# print(good_matches)
# for m in good_matches:
#   print(m.queryIdx) # 2, 6, 8, 9, ...
#   print(kpsL[m.queryIdx]) # <KeyPoint 0x7f936a537e70>, <KeyPoint 0x7f936a537b70>, ...
#   print(dir(kpsL[m.queryIdx])) # 'angle', 'class_id', 'convert', 'octave', 'overlap', 'pt', 'response', 'size'
#   print(kpsL[m.queryIdx].pt) # (381.0, 331.0), (470.0, 196.0), ...

```
