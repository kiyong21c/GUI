# SVM(Support Vector Machine)

> 크게 SVC(분류,Classification), SVR(회귀, Regression) 로 나눈다.

 - SVC는 일반적으로 이진분류(Class를 0 또는 1 등으로 나눈), 다중클래스분류도 가능하다(1:N)

  -  SVC에서 사용할 수 있는 분류 모델

    1. LinearSVC : 커널트릭 미지원(하지만, 3차원 공간에서 경계면으로 분류가능)

    2. SVC : 커널트릭 지원(poly, rbf 등)

 - SVR은 마진 안으로 적당히 많은 데이터값이 들어오도록 한다.
  - SVR에서 사용할 수 있는 분류 모델

    1. LinearSVR
    
    2. SVR

### 2가지 클래스 분류/회귀

```
# 두가지 선형 분류/회귀 모델
from sklearn.linear_model import LogisticRegression # 대표적 이진회귀 알고리즘인 로지스틱회귀 모델
from sklearn.svm import LinearSVC # 대표적 이진분류 알고리즘 선형서포트벡터분류 모델
import matplotlib.pyplot as plt
import seaborn as sns
# 결정경계 시각화
from mlxtend.plotting import plot_decision_regions

# 데이터 가져오기
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=2, random_state=8)

# 모델정의
model1 = LinearSVC() # 2진 이상의 분류도 가능
model2 = LogisticRegression() # 2진 이상의 분류도 가능


plt.figure(figsize=(12, 5))
for idx, (model, title) in enumerate(zip([model1, model2], ["LinearSVC", "LogisticRegression"])):
  # 학습
  clf = model.fit(X, y)
  # 시각화
  plt.subplot(1,2, idx+1)
  sns.scatterplot(X[:, 0], X[:, 1], hue=y) # 각각의 파라미터(Data)는 1D 이어야 한다 : (N, )
  # 시각화(2D 경계 시각화 모듈 활용)
  plot_decision_regions(X, y, clf=model, legend=2) # clf파라미터에는 적용할 "모델"을 넣는다
  plt.title(title)
  plt.xticks([])
  plt.yticks([])
  plt.xlabel("w[0]")
  plt.ylabel("w[1]")

plt.show()
```

### 선형으로 구분되지 않는 클래스를 가진 데이터

```
# 두가지 선형 분류/회귀 모델
from sklearn.linear_model import LogisticRegression # 대표적 이진회귀 알고리즘인 로지스틱회귀 모델
from sklearn.svm import LinearSVC # 대표적 이진분류 알고리즘 선형서포트벡터분류 모델
import matplotlib.pyplot as plt
import seaborn as sns
# 결정경계 시각화
from mlxtend.plotting import plot_decision_regions

# 데이터 가져오기
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state=8)
y = y%2 # 0,1,2,3 → 0,1

# 모델생성 및 학습
clf = LinearSVC().fit(X, y)

# 시각화
sns.scatterplot(X[:, 0], X[:, 1], hue=y)
# 시각화(2D 경계 시각화 모듈 활용)
plot_decision_regions(X, y, clf=clf, legend=2) # clf파라미터에는 적용할 "모델 또는 학습된모델"을 넣는다
plt.xlabel("attr 0")
plt.ylabel("attr 1")
plt.show()
```

이러한 데이터셋에는 선형 분류 모델은 적합하지 않다

→ 특성을 제곱한 특성 축을 추가해 차원을 확장해보자

```
# 두가지 선형 분류/회귀 모델
from sklearn.linear_model import LogisticRegression # 대표적 이진회귀 알고리즘인 로지스틱회귀 모델
from sklearn.svm import LinearSVC # 대표적 이진분류 알고리즘 선형서포트벡터분류 모델
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# 결정경계 시각화
from mlxtend.plotting import plot_decision_regions

# 데이터 가져오기
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state=8)
y = y%2 # 0,1,2,3 → 0,1

# 두번째 특성을 제곱하여 축을 추가(2 → 3차원)
# X데이터 자체의 차원(특성)을 추가하는 방법(또는, 데이터의 축 각각을 x, y, z변수로 만드는 방법도 있음)
# X데이터의 차원(특성)을 추가(컬럼 가로쌓기)하는 두가지 함수
X_new = np.c_[X, X[:, 1] ** 2] # np.c_[] 사용시 파라미터간의 차원 달라도 됨
# X_new = np.hstack([X, X[:, 1:] ** 2]) # np.hstack() 사용시 파라미터간의 차원이 동일해야함

# 3차원 그래프 그리기
from mpl_toolkits.mplot3d import Axes3D
plt.figure(figsize=(10, 5))
ax_3d = plt.subplot(1,2,1, projection='3d', elev=-152, azim=-26) # elev : 위/아래, azim : 좌/우
# ax_3d = fig.add_subplot(1,2,1, projection='3d') : 이렇게도 가능
# y==0 클래스 그리기
ax_3d.scatter(X_new[y==0, 0], X_new[y==0, 1], X_new[y==0, 2])
# y==1 클래스 그리기
ax_3d.scatter(X_new[y==1, 0], X_new[y==1, 1], X_new[y==1, 2])

plt.xlabel("attr 0")
plt.ylabel("attr 1")
ax_3d.set_zlabel("attr 1^2") # z축의 경우 plt.zlabel() 사용불가 → 3차원축.set_zlabe() 해야함
# plt.zlabel("attr 1^2")
```

# LinearSVC 모델을 통해 평면 결정 경계 시각화

```
# 두가지 선형 분류/회귀 모델
from sklearn.linear_model import LogisticRegression # 대표적 이진회귀 알고리즘인 로지스틱회귀 모델
from sklearn.svm import LinearSVC # 대표적 이진분류 알고리즘 선형서포트벡터분류 모델
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# 결정경계 시각화(1D/2D)
from mlxtend.plotting import plot_decision_regions

# 데이터 가져오기
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state=8)
y = y%2 # 0,1,2,3 → 0,1

# 두번째 특성을 제곱하여 축을 추가(2 → 3차원)
# X데이터 자체의 차원(특성)을 추가하는 방법(또는, 데이터의 축 각각을 x, y, z변수로 만드는 방법도 있음)
# X데이터의 차원(특성)을 추가(컬럼 가로쌓기)하는 두가지 함수
X_new = np.c_[X, X[:, 1] ** 2] # np.c_[] 사용시 파라미터간의 차원 달라도 됨
# X_new = np.hstack([X, X[:, 1:] ** 2]) # np.hstack() 사용시 파라미터간의 차원이 동일해야함

# 모델생성 및 학습
clf = LinearSVC().fit(X_new, y)
#clf.coef_ # array([[ 0.11516633, -0.19519795, -0.05529015]]) : 3개의 특징 각각의 w(기울기/가중치)
# clf.intercept_ # array([1.15414126]) : bias(절편)
coef, intercept = clf.coef_.ravel(), clf.intercept_ # ravel() 차원을 하나 낮춰줌(2 → 1차원)
# coef # array([ 0.11589501, -0.1942968 , -0.05539527])

# 3차원 그래프 축 그리기
from mpl_toolkits.mplot3d import Axes3D
plt.figure(figsize=(10, 5))
ax_3d = plt.subplot(1,2,1, projection='3d', elev=-152, azim=-26) # elev : 위/아래, azim : 좌/우
# ax_3d = fig.add_subplot(1,2,1, projection='3d') : 이렇게도 가능

# 선형 결정 경계 그리기
xx = np.linspace(X_new[:, 0].min(), X_new[:, 0].max(), 50) # attr0 축을 50개의 값으로 채운다
yy = np.linspace(X_new[:, 1].min(), X_new[:, 1].max(), 50) # attr1 축을 50개의 값으로 채운다
XX, YY = np.meshgrid(xx, yy) # XX : attr0축을 나눈 값들을 attr1축으로 쌓는다
ZZ = (coef[0]*XX + coef[1]*YY + intercept)/-coef[2] # XX → x1, YY → x2, ZZ → x3 : XX,YY,ZZ각각은 특성을 나타내는 그리드 좌표
ax_3d.plot_surface(XX, YY, ZZ) # 3차원 축이 생성되고 나면, plot_surface() 사용가능

# y==0 클래스 그리기
ax_3d.scatter(X_new[y==0, 0], X_new[y==0, 1], X_new[y==0, 2])
# y==1 클래스 그리기
ax_3d.scatter(X_new[y==1, 0], X_new[y==1, 1], X_new[y==1, 2])

plt.xlabel("attr 0")
plt.ylabel("attr 1")
ax_3d.set_zlabel("attr 1^2") # z축의 경우 plt.zlabel() 사용불가 → 3차원축.set_zlabe() 해야함
# plt.zlabel("attr 1^2")
```

```
# 두가지 선형 분류/회귀 모델
from sklearn.linear_model import LogisticRegression # 대표적 이진회귀 알고리즘인 로지스틱회귀 모델
from sklearn.svm import SVC # 대표적 이진분류 알고리즘 서포트벡터분류 모델(커널 적용가능한 모델)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# 결정경계 시각화(1D/2D)
from mlxtend.plotting import plot_decision_regions

# 데이터 가져오기
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:, :2]
iris.target[iris.target == 2] = 1 # target을 0, 1 두가지로만 구분
y = iris.target

# 모델생성 및 학습
clf = SVC(kernel='rbf', C=1, gamma=0.3).fit(X, y)
# 결정경계 시각화
plot_decision_regions(X, y, clf=clf)

sns.scatterplot(X[:, 0], X[:, 1], hue=y, s=1)

# 서포트 벡터
sv = clf.support_vectors_ # 서포트 벡터의 좌표들
# dual_coef_의 부호에 의해 서포트 벡터의 클래스 레이블이 결정됨(클래스를 가르는 서포트 벡터)
sv_labels = clf.dual_coef_.ravel() > 0
sns.scatterplot(sv[:, 0], sv[:, 1], hue=sv_labels, s=10)
```
