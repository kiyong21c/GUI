## 남자 데이터 정의

```python
import pandas as pd
file_name = '/Users/kiyongseo/PythonDataWorkspace/프로젝트/202108_202108_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name,
                     skiprows=3,
                     index_col='행정기관', # read_excel, read_csv 등 파일 읽어 올때 index 컬럼을 지정하여 가져오기
                     usecols='B,E:Y')    # 필요한 열만 가져오기(B, E부터 Y까지 열)
df_m.head(3)
```

```python
df_m.iloc[0] # , 가 없어야 숫자로 사용가능
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,195,951(str) → 1195951(str) → 1195951(int)
df_m
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name,
                     skiprows=3,
                     index_col='행정기관', # read_excel, read_csv 등 파일 읽어 올때 index 컬럼을 지정하여 가져오기
                     usecols='B, AB:AV')    # 필요한 열만 가져오기(AB부터 AV까지 열)
df_w.head(3)
# 0~4세.1 : 컬럼 인덱스 호출시 중복될것을 미연에 방지하기 위하여 .1, .2등을 자동으로 붙여줌
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼명 통일해야 이후 시각화 등 자료활용이 용이함
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용 시, 마이너스 글자가 깨지는 현상 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2021 대한민국 인구 피라미드')

plt.savefig('2021_인구피라미드.png', dpi=100)
plt.show() # show()가 savefig()보다 앞에 나오게 되면 저장이 안됨
```
