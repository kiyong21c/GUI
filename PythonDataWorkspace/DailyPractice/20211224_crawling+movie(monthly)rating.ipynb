```
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime

print("---------영화 인기 순위(월간) -----------")
year = int(input("찾으실 '연도'를 입력해주세요 :")) # input 사용시 기본이 string 타입
month = int(input("찾으실 '월'을 입력해주세요 :"))


month_start = int(str(datetime(year, month, 1).date()).replace("-","")) # datetime(연,월,일).date() -> str() 지정시 0000-00-00의 string 형태로 설정됨
month_last = int(str(datetime(year, month+1, 1).date()).replace("-",""))


date = list(range(month_start, month_last)) # range()함수의 인자로 string 타입이 들어가서는 안됨

titles = [] # 빈 리스트를 만들때는 반복문이 영향을 미치지 않도록
points = []

for time in date:
  url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={}".format(date)
  res = requests.get(url)
  res.raise_for_status()  # 문제있으면 프로그램 종료
  soup = BeautifulSoup(res.text, "lxml")


  title_list = soup.find_all("td", attrs={"class":"title"})
  for title in title_list:
    titles.append(title.a.get_text())

  point_list = soup.find_all("td", attrs={"class":"point"})
  for point in point_list:
    points.append(point.get_text())


df = pd.DataFrame({'name':titles, 'point': points})
df
```
