# 4. 스타일

```python
import matplotlib.pyplot as plt

import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 마이너스 문자가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth=10)
```

```python
plt.plot(x, y, linewidth=5)
```

## Marker

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

## 색깔

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g') # matplotlib에서 미리 정의된 이름
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:') # color, marker, linestyle
```

```python
plt.plot(x, y, 'go') # color, marker, (linestyle)
# plt.plot(x, y, color='g', marker='o', linestyle='None') 이것과 동일
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', mec='blue', ls=':')
# makerfacecolor → mfc
# markersize → ms
# markeredgecolor → mec
# linestyle → ls
```

## 투명도

```python
plt.plot(x, y, color='g', marker='o', ms=10, alpha=0.3) # alpha : 투명도(0 ~ 10)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5)) # 가로, 세로
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=200) # dpi(dots per inch) : 해상도(확대한 개념정도로 이해)
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff') # google에서 color picker 검색이용
plt.plot(x, y)
```
