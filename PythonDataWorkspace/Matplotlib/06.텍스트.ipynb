# 6. 텍스트

```python
import matplotlib.pyplot as plt

import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 마이너스 문자가 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker='o')
for idx, text in enumerate(y): # y 리스트의 인덱스와 값을 추출
    plt.text(x[idx], y[idx]+0.3, text, ha='center', color='blue') # horizontal alignment → ha : text의 x점 기준 가운데 정렬
```
