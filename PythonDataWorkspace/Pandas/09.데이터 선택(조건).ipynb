# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 키 column의 값이 185이상인지 각 row별로 True/False
```

```python
filt = df['키'] >= 185
df[filt] # filt조건에 해당하는 row 가져옴
```

```python
df[~filt] # filt조건을 역으로 적용 : ~filt
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학']  # df.loc[row_select, col_select]
# df['키'] >= 185 → row 조건
```

```python
df.loc[df['키'] >= 185, ['이름','수학','과학']]
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185이상인 북산고 학교 학생 데이터
# 조건이 두개 이상일때 & 사용하면서 각 조건을 소괄호()로 묶어줘야 에러가 안난다
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

###  str 함수

```python
filt = df['이름'].str.startswith('송') # 송씨 성을 가진 사람
# .str : 문자열로 변경
# .startswith(n) : n문자로 시작하는 문자열
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 태가 들어가는 사람
# .contains(n) : n문자가 포함되는 문자열
df[filt]
```

```python
df[~filt] # 이름에 태가 안 들어가는 사람
```

```python
langs = ['Python','Java']
filt = df['SW특기'].isin(langs) # SW특기가 Python 이거나 Java 인 사람
df[filt]
```

```python
df # Python, python, Java, java, Javascript를 효과적으로 선택하기
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs) # filt는 True/False를 row 별로 나타냄
df[filt]
```

```python
df['SW특기'].str.contains('Java') # True/False/Nan 조건으로 나타낼 수 있지만
```

```python
filt = df['SW특기'].str.contains('Java') 
df[filt] # NaN이 있으면 처리를 못함
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터를 True로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False) 
df[filt]
```
